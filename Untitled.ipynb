```python
# import libraries 

import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
from math import *
import glob
import scipy.io
from sklearn.utils import shuffle
```

```python
# Initialize parameters and read data

RANDOM_SEED = 1
length = 64
nlinks = 9
output_nodes = length*2*nlinks
input_nodes = length*2*nlinks
featureSize = length*2*nlinks
hidden_nodes1 = 400
hidden_nodes2 = 700
ncsi = 50
Ntrain = 40
Ntest = 7
dataPath = 'C:/Users/origingpu/Desktop/ReverberationChamber/bglearn/Data/'

X = np.zeros(((Ntrain+Ntest)*ncsi,featureSize))
Y = np.zeros(((Ntrain+Ntest)*ncsi,featureSize))
Count = 0         
for file in glob.glob(dataPath+'CSI*_LOC_A_DIRECTION_north_*deepika_regani*_truncated.mat'):  
    if  "background" in file: 
        continue
    if (Count<Ntrain+Ntest):
      #  print(file)
        mat = scipy.io.loadmat(file)
        data = mat['CSI_mtx'][:,:,:,0:ncsi]
        data = data.reshape(length*nlinks,ncsi)
        firstname = file.split("_B9.mat_truncated.mat",1)[0]
        filebg = firstname+ "_background_B9.mat_truncated.mat"       
        matbg = scipy.io.loadmat(filebg)
        databg = matbg['CSI_mtx'][:,:,:,0:ncsi]
        databg = databg.reshape(length*nlinks,ncsi)
        feature = np.concatenate([data,data])        
        feature[0::2]=np.real(data)
        feature[1::2]=np.imag(data) 
        featurebg = np.concatenate([databg,databg])        
        featurebg[0::2]=np.real(databg)
        featurebg[1::2]=np.imag(databg)   
        X[(Count)*ncsi:(Count+1)*ncsi,:] = np.real(feature.transpose())        
        Y[(Count)*ncsi:(Count+1)*ncsi,:] = np.real(featurebg.transpose())
        Count = Count + 1 
        
#X, Y = shuffle(X, Y, random_state=0)
        
X_train = X[0:Ntrain*ncsi,:]
y_train = Y[0:Ntrain*ncsi,:]
X_test = X[Ntrain*ncsi:(Ntrain+Ntest)*ncsi,:]
y_test = Y[Ntrain*ncsi:(Ntrain+Ntest)*ncsi,:]
     
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)



```

```python
from collections import namedtuple
def get_batches(x,y,batch_size):
    for i in range(0,x.shape[0],batch_size):
        batchx = x[i:i+batch_size+1]
        batchy = y[i:i+batch_size+1]
        yield (batchx, batchy)

def fc_layer(inputs,channels_in,channels_out,act,bias,name="hidden_layer"):
    if act=='elu':
        activation=tf.nn.elu
    elif act=='tanh':
        activation=tf.nn.tanh
    elif act=='relu':
        activation=tf.nn.relu
    elif act=='sigmoid':
        activation=tf.nn.sigmoid
    elif act=='none':
        activation=lambda x:x
    W=tf.Variable(tf.truncated_normal(dtype=tf.float32,shape=(channels_in,channels_out),stddev=0.1),name='W')
    b=tf.Variable(tf.zeros(channels_out),name='b')
    if(bias==1):
        h=activation(tf.add(tf.matmul(inputs,W),b))
    elif(bias==0):
        h=activation(tf.matmul(inputs,W))
    return h

def build_nn(num_nodes,num_layers,bias,activation_unit,lr,x_size,y_size,epochs,batch_size,file_writer,log_string):    
    tf.reset_default_graph()


    x = tf.placeholder(tf.float32, [None, x_size ],name="x")
    y = tf.placeholder(tf.float32, [None, y_size],name="y")

    with tf.name_scope("Neural_Network"):
        h=fc_layer(x,x_size,num_nodes,activation_unit,bias,name="hidden_layer_1")
        for i in range(2,num_layers+1):
            h=fc_layer(h,num_nodes,num_nodes,activation_unit,bias,name="hidden_layer_"+str(i))

    out=fc_layer(h,num_nodes,y_size,act='none',bias=1,name="output_layer")
    #with tf.name_scope('logit'):
    #    pred=tf.argmax(out,name='prediction')

#     embedding_size=400
#     embedding_input=h
#     embedding = tf.Variable(tf.zeros([x_test_size, embedding_size]), name="test_embedding")
#     assignment = embedding.assign(embedding_input)


    with tf.name_scope('cost'):
#         loss=tf.nn.softmax_cross_entropy_with_logits(labels=y,x_size logits=out,name='loss')
        cost = tf.reduce_mean(tf.abs(out-y))
        #cost = tf.nn.l2_loss(out-y)
        tf.summary.scalar('cost',cost)

#     with tf.name_scope("Accuracy"):
#         correct_prediction = tf.reduce_mean(tf.abs(out-y),name='correct_prediction')
#         accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32), name='accuracy')
#         tf.summary.scalar('accuracy',accuracy)

    with tf.name_scope('train'):
        optimizer = tf.train.AdamOptimizer(learning_rate=lr).minimize(cost)

    merged=tf.summary.merge_all()

#     export_nodes=['x','y','cost','optimizer','merged','correct_prediction','accuracy','embedding','assignment']
#     Graph=namedtuple('Graph',export_nodes)
#     local_dict=locals()
#     graph=Graph(*[local_dict[each] for each in export_nodes])

#     return graph
    with tf.Session() as sess:
        init = tf.global_variables_initializer()
        sess.run(init)        
        total_batch = int(X_train.shape[0]/batch_size)
        for epoch in range(epochs):
            avg_cost = 0.
            # Loop over all batches
            for b,(batch_x,batch_y) in enumerate(get_batches(X_train,y_train,batch_size),1):
                iteration=(epoch*total_batch)+b               
                c,_ = sess.run([cost,optimizer], feed_dict={x: batch_x,y: batch_y})
                # Compute average loss
                avg_cost += c / total_batch
            # Display logs per epoch step
            if epoch % display_step == 0:
#                 sess.run(model.assignment,feed_dict={model.x: X_test, model.y:y_test})

                tc,out_vec=sess.run([cost,out],feed_dict={x: X_test,y:y_test})
                #file_writer.add_summary(summ,iteration)
                print("Epoch:", "%04d" % (epoch+1), "train_cost=","{:.9f}".format(c), "test_cost=", "{:.9f}".format(tc))
                
        #for i in range(1,batch_size+1):
            t = np.arange(0.,128, 1)
        #figure(1)
            fig_size = np.zeros((2,1))
            fig_size[0]=15
            fig_size[1] = 10
            plt.rcParams["figure.figsize"] = fig_size
        for i in range(1):
            #plt.subplot(batch_size/4,4,i)
            plt.plot(t,np.abs(out_vec[i,0:128]),'r--',t,np.abs(y_test[i,0:128]),'g--')
        plt.legend(['Predicted','Original'])        
        plt.show()
        plt.close()        
        print("Optimization Finished!")
      

#         return (sess.run(model.pred,feed_dict={model.x: X_test, model.y:y_test}))       


```

```python
training_epochs=400
display_step=50
batch_size=16
num_nodes=400
x_size = input_nodes
y_size = output_nodes
for num_layers in [4]:
        for bias in [1]:
            for act in ['relu']:
                for lr in [0.0001]:                    
                    log_string= 'logs/various_architectures/layers={},bias={},act={},lr={}'.format(num_layers,bias,act,lr)
                    file_writer=tf.summary.FileWriter(log_string)
                    build_nn(num_nodes,num_layers,bias,act,lr,x_size,y_size,training_epochs,batch_size,file_writer,log_string)
```
